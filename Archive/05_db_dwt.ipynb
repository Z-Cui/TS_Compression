```python
import pywt
import pywt.data
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [5, 2]

x = pywt.data.ecg()
x.shape
```

```python
A1, _ = pywt.dwt(x,'db20')
A2, _ = pywt.dwt(A1,'db20')
# A = A2[0:256]
B1 = pywt.idwt(A2, None, 'db20)
B2 = pywt.idwt(B1, None, 'db20')
float(np.linalg.norm(B2[0:1024] - x))
```
