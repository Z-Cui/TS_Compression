```python
import numpy as np
import math
from tslearn.preprocessing import TimeSeriesScalerMeanVariance
from tslearn.piecewise import PiecewiseAggregateApproximation
from tslearn.piecewise import SymbolicAggregateApproximation, \
    OneD_SymbolicAggregateApproximation

filename = 'synthetic_size50k_len256_znorm.bin'
original50k = np.fromfile(filename, dtype=np.float32).reshape(-1,256)
```

```python
x = original50k
# SAX transform
n_sax_symbols = 256
sax_128 = SymbolicAggregateApproximation(n_segments=128, alphabet_size_avg=n_sax_symbols)
sum128_x = sax_128.fit_transform(x)
rec128_x = sax_128.inverse_transform(sum128_x)

sax_64 = SymbolicAggregateApproximation(n_segments=64, alphabet_size_avg=n_sax_symbols)
sum64_x = sax_64.fit_transform(x)
rec64_x = sax_64.inverse_transform(sum64_x)

sax_32 = SymbolicAggregateApproximation(n_segments=32, alphabet_size_avg=n_sax_symbols)
sum32_x = sax_32.fit_transform(x)
rec32_x = sax_32.inverse_transform(sum32_x)

print(x.shape, sum128_x.shape, sum64_x.shape, sum32_x.shape)
print(x.shape, rec128_x.shape, rec64_x.shape, rec32_x.shape)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [20, 7]
i = 2
plt.subplot(2, 4, 1)
plt.title("original 256 float32")
plt.plot(x[i])
plt.subplot(2, 4, 2)
plt.title("compressed 128 SAX")
plt.plot(sum128_x[i])
plt.subplot(2, 4, 3)
plt.title("compressed 64 SAX")
plt.plot(sum64_x[i])
plt.subplot(2, 4, 4)
plt.title("compressed 32 SAX")
plt.plot(sum32_x[i])
plt.subplot(2, 4, 5)
plt.title("original 256 float32")
plt.plot(x[i])
plt.subplot(2, 4, 6)
plt.title("reconstructed 128 SAX")
plt.plot(rec128_x[i])
plt.subplot(2, 4, 7)
plt.title("reconstructed 64 SAX")
plt.plot(rec64_x[i])
plt.subplot(2, 4, 8)
plt.title("reconstructed 32 SAX")
plt.plot(rec32_x[i])
```

```python
# Calculate RMSE
from sklearn.metrics import mean_squared_error
from math import sqrt
def rmse(x,y):
    return sqrt(mean_squared_error(x, y))
def summarize_data(x):
    print("Average:", np.mean(x), " Mean:", np.median(x), " Max:", np.max(x), " Min:", np.min(x), "\n")
```

```python
errorList_128_synthetic = []
errorList_64_synthetic = []
errorList_32_synthetic = []
for i in range(50000):
    errorList_128_synthetic.append(rmse(x[i], rec128_x[i]))
    errorList_64_synthetic.append(rmse(x[i], rec64_x[i]))   
    errorList_32_synthetic.append(rmse(x[i], rec32_x[i]))
```

```python
# Reconstruction error for 50k synthetic time_series
print("Reconstruction Error of 50k synthetic time series: 128 -> 256")
summarize_data(errorList_128_synthetic)
print("Reconstruction Error of 50k synthetic time series: 64 -> 256")
summarize_data(errorList_64_synthetic)
print("Reconstruction Error of 50k synthetic time series: 32 -> 256")
summarize_data(errorList_32_synthetic)
```
